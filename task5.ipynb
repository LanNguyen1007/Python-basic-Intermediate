```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import re 
```

```python
df1 = pd.read_csv('Ecommerce_Purchases.csv')

col_name5 = ['CC Provider', 'AM or PM', 'Purchase Price']
data_task5 = df1[col_name5]

data_task5.head(10)
```

```python
grouped = data_task5.groupby(['CC Provider', 'AM or PM'])['Purchase Price'].sum()

grouped = grouped.unstack()
```

```python
grouped.plot(kind='bar', figsize=(13,6))

plt.xlabel('CC Provider', size=15)
plt.ylabel('Purchase Price', size=15)
plt.legend()
```
