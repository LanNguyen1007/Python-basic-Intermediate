```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import re 

df1 = pd.read_csv('Ecommerce_Purchases.csv')
```

```python
def extract_title(name):
    p = re.compile(r"\s+([A-Z]{2})\s+")
    return p.search(name).groups(2)[0].strip()
df1['Address'] = df1['Address'].apply(lambda name: extract_title(name))
```

```python
col_name_4 = ['Address', 'AM or PM', 'Purchase Price']
data_task4 = df1[col_name_4]
data_task4
```

```python
grouped = data_task4.groupby(['Address', 'AM or PM'])['Purchase Price'].sum() 

grouped = grouped.unstack()
grouped.plot(kind='line', figsize=(15, 8))

plt.ylabel('Purchase Price',size=15)
plt.xlabel('State',size=15)
plt.legend()

```

```python
grouped = data_task4.groupby(['Address', 'AM or PM'])['Purchase Price'].sum()
grouped = grouped.unstack()

grouped.plot(kind='bar', figsize=(15, 8))

plt.ylabel('Purchase Price',size=15)
plt.xlabel('State',size=15)
plt.legend()
```
