```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Ecommerce_Purchases.csv')
```

```python
col_names =['Browser Info','Job', 'Purchase Price']

Browsers = df1[col_names]

Browsers.head(5)
```

```python
Browsers_purchase_total = Browsers.groupby(['Job', 'Browser Info'])['Purchase Price'].sum().reset_index(name='Total')
Browsers_purchase_total.head(10) 
```

```python
Browsers_buyer = Browsers_purchase_total.sort_values(by='Total', ascending=False)
Browsers_buyer.head(10)
```

```python
Browsers_chart = Browsers_buyer.head(10) 

plt.figure(figsize=(12, 6))  
plt.scatter(Browsers_chart['Browser Info'].str.split('/').str[0], Browsers_chart['Job'], c=Browsers_chart['Total'], cmap='viridis', s =Browsers_chart['Total']*5)
plt.colorbar(label='Purchase Price')
plt.xlabel('Internet Browser')
plt.ylabel('Job')
plt.title('Purchase Value vs. Internet Browser and Job')
```
