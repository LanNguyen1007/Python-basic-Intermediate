```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Ecommerce_Purchases.csv')

col_names = ['Job', 'Purchase Price']

jobs = df1[col_names]

jobs
```

```python
jobs_price_mean = jobs.groupby('Job')['Purchase Price'].mean().reset_index(name='Price Mean') 

jobs_top10_mean = jobs_price_mean.sort_values(by='Price Mean', ascending=False).head(10)

jobs_top10_mean
```

```python
jobs_top10_mean.plot(x='Job', y='Price Mean', figsize=(15,7), kind='barh')

plt.xlabel('Price Mean', size=15)
plt.ylabel('Job', size=15)
plt.title('Top 10 Job with the highest Price Mean')
```
